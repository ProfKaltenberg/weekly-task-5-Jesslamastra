# Question 1

Using the dataset "us-cities-top-1k-multi-year.csv":
1. Create a line graph of total population over time for the states: NY, PA, CT, NJ (10 pts)
2. Be sure to include a title of the graph, labels on your axis, and a legend. (5 pts)
3. Specify colors to be used in your code. (5 pts)

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv(r'C:\Users\lamas\OneDrive\Documents\eco590\weekly-task-5-Jesslamastra\weekly task 5\us-cities-top-1k-multi-year.csv')
data.head() #check data implemented
```

```python
#line graph

# Filter for wanted states
select_states = data[(data['State'] == 'New York') | (data['State']== 'Connecticut') | (data['State']== 'New Jersey') | (data['State']== 'Pennsylvania')]
pop_time_state = select_states.groupby(['year', 'State'])['Population'].sum().reset_index() #cleans raw data so it can plot
pop_time_state['year'] = pop_time_state['year'].astype(int) # Convert year to int because when i plot it was giving me float on axis


colors = {
    'New York': 'purple',
    'New Jersey': 'navy',
    'Pennsylvania': 'green',
    'Connecticut': 'yellow'
} 
#created a color dictionary so each state can be a specified color
states = ['New York', 'New Jersey', 'Pennsylvania', 'Connecticut'] #creates list of states for loop 

for s in states: #loops through each state
    state_pop = pop_time_state[pop_time_state['State'] == s] #filters data to one state at a time
    plt.plot(
        state_pop['year'],
        state_pop['Population'],
        label=s,
        color=colors[s] #draws a line for the state with corresponding color in dictionary
    )


plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') #plot legend

plt.title("Population Over Time by State", fontsize=12, fontweight='bold', loc='center') #create title
#axsis lables
plt.xlabel("Year")
plt.ylabel("Total Population (in 10 million)")

plt.xticks(sorted(pop_time_state['year'].unique())) #make sure x axsis ticks are correct

plt.show() #shows graph
```
